```python
def match_collect(df, gamemode, count, season):
    match_list = []
    
    for player in df['playerId'].values:
        count += 1
        
        ### 열명의 데이터 수집 후 1분의 시간을 주어서 다시 수집하기 위해, pubg 정책
        if count % 10 == 0:
            time.sleep(60)
            
        url = 'https://api.pubg.com/shards/steam/seasons/{}/gameMode/{}/players?filter[playerIds]={}'.format(season, gamemode, player)
        header = {
            "Authorization": KEY,
            "Accept": "application/vnd.api+json"
        }
        response = requests.get(url, headers = header)
        datas = response.json()
        
        ### 솔로 매치에서의 필요한 유저아이디와 매치아이디를 분류
        for i in range(len(datas['data'][0]['relationships']['matchesSolo']['data'])):
            data = datas['data'][0]['relationships']['matchesSolo']['data'][i]['id']
            match_list.append({'user_id':player, 'match_id':data})
        print('user_id:{}, count:{}'.format(player, count))
        
    return match_list
```

### MAP
- 수집한 매치 데이터에서 미라마(사막)맵을 추출하기 위한 코드 작성
- 에란겔, 미라마, 사녹, 비켄이 4가지 맵이 있지만, 본인의 목적은 미라마(사막)맵의 데이터를 분석하기 위해서 사막맵만 분류함

```python
match_df = pd.read_csv("match_duplicates.csv")
```

```python
match_df = match_df.drop(11357)
```

```python
def map_collect(df, key):
    map_list = []
    count = 0

    for match, user in df.values:
        count += 1
        url = 'https://api.pubg.com/shards/steam/matches/{}'.format(match)
        header = {
            "Authorization": key,
            "Accept": "application/vnd.api+json"
        }
        response = requests.get(url, headers = header)
        datas = response.json()
        
        ### 맵 이름 추출
        data = datas['data']['attributes']['mapName']
        
        map_list.append({'user_id':user, 'match_id':match, 'map':data})
        
        print('match:{}, count:{}'.format(match, count))
    
    return map_list
```

- 맵 리스트를 데이터프레임으로 변환

### Location
- 유저의 첫 착률 지점 좌표를 수집하기위한 asset 데이터 수집

```python
map_df = pd.read_csv('map_second.csv')
```

```python
map_df.tail()
```

```python
def asset_collect(df, key):
    url_list = []
    count = 0

    for m, match, user in df.values:
        count += 1
        url = 'https://api.pubg.com/shards/steam/matches/{}'.format(match)
        header = {
            "Authorization": key,
            "Accept": "application/vnd.api+json"
        }
        response = requests.get(url, headers = header)
        datas = response.json()
        
        ### asset 데이터 수집
        if len(datas.keys()) == 4:
            for i in range(len(datas['included'])):
                if datas['included'][i]['type'] == 'asset':
                    link = datas["included"][i]["attributes"]["URL"]
                    url_list.append({'map':m, 'match': match, 'user': user, 'url': link})

            print('match:{}, count:{}'.format(match, count))
        
    return url_list
```

- 수집한 asset 리스트를 데이터 프레임으로 변환

```python
url_df = pd.DataFrame(asset_collect(map_df, KEY))
```

```python
url_df.to_csv('asset_second.csv', index=False)
```

```python
def lend(name, datas, match, url):
    ls = {}
    for i in range(len(datas)):
        if list(datas[i].keys())[0] == 'character':
            if datas[i]['character']['name'] == name:
                if datas[i]['_T'] == 'LogParachuteLanding':
                    ls = {
                        'match':match,
                        'url':url,
                        'name': name,
                        'x':datas[i]['character']['location']['x'],
                        'y':datas[i]['character']['location']['y'],
                    }
    return ls
```

```python
def data_collect(df, key):
    count = 1571
    
    for m, match, url, user in df[1571:].values:
        count += 1
        link = 'https://api.pubg.com/shards/steam/matches/{}'.format(match)
        headers = {
            "Authorization" : key,
            "Accept" : "application/vnd.api+json"
        }
        response = requests.get(link, headers = headers)
        datas = response.json()
        
        ### 유저 스텟 데이터 수집 ###
        df_user = []
        
        for i in range(len(datas["included"])):
            if datas["included"][i]["type"] == "participant":
                df_user.append({
                    "match" : datas["data"]["id"],
                    'DBNOs': datas["included"][i]["attributes"]["stats"]["DBNOs"],
                    'assists': datas["included"][i]["attributes"]["stats"]["assists"],
                    'boosts': datas["included"][i]["attributes"]["stats"]["boosts"],
                    'damageDealt': datas["included"][i]["attributes"]["stats"]["damageDealt"],
                    'deathType': datas["included"][i]["attributes"]["stats"]["deathType"],
                    'headshotKills': datas["included"][i]["attributes"]["stats"]["headshotKills"],
                    'heals': datas["included"][i]["attributes"]["stats"]["heals"],
                    'killPlace': datas["included"][i]["attributes"]["stats"]["killPlace"],
                    'killStreaks': datas["included"][i]["attributes"]["stats"]["killStreaks"],
                    'kills': datas["included"][i]["attributes"]["stats"]["kills"],
                    'longestKill': datas["included"][i]["attributes"]["stats"]["longestKill"],
                    'name': datas["included"][i]["attributes"]["stats"]["name"],
                    'playerId': datas["included"][i]["attributes"]["stats"]["playerId"],
                    'revives': datas["included"][i]["attributes"]["stats"]["revives"],
                    'rideDistance': datas["included"][i]["attributes"]["stats"]["rideDistance"],
                    'roadKills': datas["included"][i]["attributes"]["stats"]["roadKills"],
                    'swimDistance': datas["included"][i]["attributes"]["stats"]["swimDistance"],
                    'teamKills': datas["included"][i]["attributes"]["stats"]["teamKills"],
                    'timeSurvived': datas["included"][i]["attributes"]["stats"]["timeSurvived"],
                    'vehicleDestroys': datas["included"][i]["attributes"]["stats"]["vehicleDestroys"],
                    'walkDistance': datas["included"][i]["attributes"]["stats"]["walkDistance"],
                    'weaponsAcquired': datas["included"][i]["attributes"]["stats"]["weaponsAcquired"],
                    'winPlace': datas["included"][i]["attributes"]["stats"]["winPlace"],
                 })
        user_df = pd.DataFrame(df_user).sort_values("winPlace").reset_index(drop = True)

        ### 해당 유저의 첫 착륙 좌표 데이터 수집 ###
        url = url
        header = {
            "Authorization": KEY,
            "Accept": "application/vnd.api+json"
        }
        response = requests.get(url, headers = header)
        datas = response.json()

        point_ls = []

        for name in user_df['name']:
            point_ls.append(lend(name, datas, match, url))
        point_df = pd.DataFrame(point_ls)
        
        ### 유저 스텟과 유저의 첫 착륙 좌표 병합 ###
        df = user_df.merge(point_df, on=['name', 'match'])
        df.to_csv("datas2/location_{}.csv".format(count), index=False)
        
        
        ### 비행기 경로 좌표 수집 ###
        ls_postion = []
        for i in range(len(datas)):
            if list(datas[i].keys())[0] == "character":
                if datas[i]["character"]["name"] == user_df['name'][0]:
                    ls_postion.append({
                        "X" : datas[i]["character"]["location"]["x"],
                        "Y" : datas[i]["character"]["location"]["y"],  
                        "Z" : datas[i]["character"]["location"]["z"],
                    })
        df1 = pd.DataFrame(ls_postion)
        
        z_key = True
        ls = []
        for z in range(100):
            if z_key: 
                if df1["Z"][z] > 150000:
                    z_key = False
            else:
                if df1["Z"][z] < 150000:
                    break
                ls.append({
                    "X" : df1["X"][z],
                    "Y" : df1["Y"][z],
                    "Z" : df1["Z"][z],
                }) 
        airplane = pd.DataFrame(ls)
        airplane.to_csv('datas2/airplane_{}.csv'.format(count), index=False)
    
        print("count:{}".format(count))
```
